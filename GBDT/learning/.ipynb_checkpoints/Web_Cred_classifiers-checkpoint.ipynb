```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import os
from sklearn.utils import class_weight
from functools import partial
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score as recall
from sklearn.metrics import precision_score as precision
from sklearn.metrics import f1_score as f1
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import cohen_kappa_score as kappa_score
from sklearn.metrics import classification_report
import pickle
```

```python
import csv
X = []
with open('/Users/apple/Academics/MS/WEBCred/data/dump/webcred_features_expanded.csv','rb') as f:
#     data = f.readlines()
    data = csv.reader(f, delimiter=',')
    for line in data:
        X.append(line)

        
tag = X[0].index('genre')
Y = [i[tag] for i in X]
# print len(Y)
# print len(X[1])
X = [row[:tag]+row[tag+1:] for row in X]
# print len(X[1])
features = X[0]
print 'Features'
print features
X = X[1:]
Y = Y[1:]
        
X_2 = []
for row in X:
    temp = []
    for item in row:
        if item is not '':
            item = float(item)
        else:
            item = -1.0
        temp.append(item)
    X_2.append(temp)
    
X = X_2

print len(X[0])

print len(X[1])
print type(X[1][2])

X = np.asarray(X)
Y = np.asarray(Y)
Y = np.ravel(Y)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)

print str(X_train.shape) + ' Training data dimension'
print str(Y_train.shape) + ' Training data labels'

print str(X_test.shape) + ' Test data dimension'
print str(Y_test.shape) + ' Test data labels'            
```

```python
classes = list(set(Y_train))
class_name =  ['article', 'broken_link', 'public_portrayals_companies_and_institutions', 'private_portrayal_personal_homepage', 'downloads', 'help', 'discussion', 'shop', 'link_collection', 'other']
class_counts = [0]*len(classes)
for i in Y_train:
    class_counts[int(i)] += 1
# cla = dict(Y_train)
print [(class_name[i], class_counts[i]) for i in range(len(classes))]
# print class_counts
```

```python
def stats(y_test, y_pred):
    
    #     print 'Accuracy:', accuracy_score(y_test, weighted_prediction)
    print 'F1 score:', f1(y_test, y_pred,average='weighted')
    print 'Recall:', recall(y_test, y_pred,
                                  average='weighted')
    print 'Precision:', precision(y_test, y_pred,
                                        average='weighted')
#     print y_test, y_pred
#     print 'Mean Squared Error:', mse(y_test, y_pred)
    print 'Kappa Score:', kappa_score(y_test, y_pred)   
    print '\n clasification report:\n', classification_report(y_test, y_pred)
    print '\n confussion matrix:\n',confusion_matrix(y_test, y_pred)

```

```python
# LR model
from sklearn.linear_model import LogisticRegression

def train_lr(X_train,Y_train, X_test,Y_test):
    lr = LogisticRegression()


    lr.fit(X_train,Y_train)

    print str(lr.score(X_train,Y_train)) + ' Training Accuracy'
    print str(lr.score(X_test,Y_test)) + ' Testing Accuracy'

    y_pred = lr.predict(X_test)
#     print 'confusion matrix'
#     print confusion_matrix(Y_test,Y_pred_lr)
    stats(Y_test, y_pred)
```

```python
# SVM model
from sklearn.svm import SVC

def train_svm(x,y,x_test,y_test):
    
    svm = SVC(class_weight='balanced',C=1e10)
    svm.fit(x,y)

    print str(svm.score(x,y)) + ' Training Accuracy'
    print str(svm.score(x_test,y_test)) + ' Testing Accuracy'

    y_pred = svm.predict(x_test)

    stats(y_test, y_pred)
# biased learning, most of thr predictions were articles, 
# as it has the maximum count in training set
```

```python
# NN
from sklearn.neural_network import MLPClassifier

def train_mlp(X_train,Y_train, X_test,Y_test):
#     print 'nodes= 128,32'
    nodes = [128,32]
    print 'nodes=', nodes
    mlp = MLPClassifier(hidden_layer_sizes=nodes)

    mlp.fit(X_train,Y_train)

    print str(mlp.score(X_train,Y_train)) + ' Training Accuracy'
    print str(mlp.score(X_test,Y_test)) + ' Testing Accuracy'

    y_pred = mlp.predict(X_test)
    stats(Y_test, y_pred)
#     print 'confusion matrix'
#     print confusion_matrix(Y_test,Y_pred_mlp)
```

```python
# decision trees
from sklearn.ensemble import GradientBoostingClassifier

def train_gbdt(X_train,Y_train,X_test,Y_test ):

    gbdt = GradientBoostingClassifier(n_estimators=100,max_depth=3,verbose=1)
    gbdt.fit(X_train,Y_train)

    print str(gbdt.score(X_train,Y_train)) + ' Training Accuracy'
    print str(gbdt.score(X_test,Y_test)) + ' Testing Accuracy'

    y_pred = gbdt.predict(X_test)
    stats(Y_test, y_pred)
#     print 'confusion matrix'
#     print confusion_matrix(Y_test,Y_pred_gbdt)
```

```python
# under sampling
from imblearn.under_sampling import RandomUnderSampler
# cluster sample
# cc = ClusterCentroids(random_state=0)

# random sample 
cc = RandomUnderSampler(random_state=0)
X_train_rus, Y_train_rus = cc.fit_sample(X_train, Y_train)
print str(X_train_rus.shape) + ' Training data dimension'
print str(Y_train_rus.shape) + ' Training labels'
```

```python
# over sampling
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2,k_neighbors=1)
X_train_ros, Y_train_ros = sm.fit_sample(X_train, Y_train)
print 'After Oversampling'
print str(X_train_ros.shape) + ' Training data dimension'
print str(Y_train_ros.shape) + ' Training labels'
```

```python
print 'svm'
print 'training for random under sampled data'
train_svm(X_train_rus, Y_train_rus, X_test, Y_test)
print
print 
print 'training for random over sampled data'
train_svm(X_train_ros, Y_train_ros, X_test, Y_test)
print
```

```python
print 'LR'
print 'training for random under sampled data'
train_lr(X_train_rus, Y_train_rus, X_test, Y_test)
print
print 
print 'training for over sampled data'
train_lr(X_train_ros, Y_train_ros, X_test, Y_test)
print
```
